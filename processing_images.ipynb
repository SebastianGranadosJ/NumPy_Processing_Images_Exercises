## Create circular mask


```python
total_rows, total_cols, total_layers = image1.shape
X, Y = np.ogrid[:total_rows, :total_cols]
print(X)
print(Y)
```

```python
center_row, center_col = total_rows/2, total_cols/2
print("Central row = ", center_row, "Central column =", center_col)
```

```python
dist_from_center = (X -center_row)**2 +(Y - center_col)**2

print(dist_from_center)
```

```python
image3 = np.array(image)
radius = (total_rows/2)**2
image3[dist_from_center > radius] = 0
plt.figure(figsize = (8 ,8))
plt.imshow(image3)
```

```python
image3 = np.array(image)

## Square filter 


side_lenght = 700
half_side = (side_lenght/2)**2
 
dist_from_center_X = (X -center_row)**2 
matrix_dist_from_center_X = np.zeros((total_rows, total_cols))  + dist_from_center_X
image3[matrix_dist_from_center_X > half_side] = 0

dist_from_center_Y = (Y -center_col)**2 
matrix_dist_from_center_Y = np.zeros((total_rows, total_cols))  + dist_from_center_Y
image3[matrix_dist_from_center_Y > half_side] = 0


plt.figure(figsize = (8 ,8))
plt.imshow(image3)
```
